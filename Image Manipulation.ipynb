## Task 2b

```python
# write a display function and call it twice for each type of image
def display_image(image):
    plt.xlabel('columns')
    plt.ylabel('rows')
    plt.imshow(image)
    plt.colorbar()
    plt.show()
    
display_image(img_png)

```

## Task 13

```python
# Display two  gamma correctied images - light and dark
gamma_light = np.clip(np.power(img_png, 1/2.2), 0, 1)
display_image(gamma_light)
```

```python
gamma_dark = np.clip(np.power(img_png, 2.2), 0, 1)
display_image(gamma_dark)
```

## Task 14

```python
# obtain the img in red, green and blue scale and horizontally stack/ join them 
red_png = img_png.copy()
red_png[:, :, [1,2]] = 0
green_png = img_png.copy()
green_png[:, :, [0,2]] = 0 
blue_png = img_png.copy()
blue_png[:,:, [0,1]] = 0
all_img = np.concatenate((red_png, green_png, blue_png), axis = 1)
plt.figure(figsize = (18,20))
plt.imshow(all_img)
```
