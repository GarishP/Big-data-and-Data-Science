## TASK 13. Counts of International flights

```python
#13A)
inUSADF = flightsAllDF.where((F.col('dest')=='United States')&(F.col('origin')!='United States')).groupby('year').sum('numflights').select(F.col('year'),F.col('sum(numflights)').alias('totalflightsin')).sort(F.col('totalflightsin').asc())
inUSADF.display()
```

```python
#13B)
outUSADF = flightsAllDF.where((F.col('dest')=='United States')&(F.col('origin')!='United States')).groupby('year').sum('numflights').select(F.col('year'),F.col('sum(numflights)').alias('totalflightsout')).sort(F.col('totalflightsout').asc())
IntflightsUSA = inUSADF.join(outUSADF, on='year', how='inner')
IntflightsUSA.withColumn("morein",IntflightsUSA['totalflightsin']>IntflightsUSA['totalflightsout']).display()
```

## TASK 14. Aggregate statistics

```python
#14A)
from pyspark.sql.functions import sum, avg, max
flightsAllDF.filter((flightsAllDF.origin!='United States')&(flightsAllDF.dest=='United States')).groupby('year').agg(sum('numflights').alias('sumflightstoUS'),avg('numflights').alias('avgflightstoUS'),max('numflights').alias('maxflighttoUS')).sort(F.col('year').asc()).display()
```
